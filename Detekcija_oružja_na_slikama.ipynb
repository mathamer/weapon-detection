```python
from google.colab import drive
import os
import zipfile

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

# Define the paths
drive_path = '/content/drive/MyDrive/NMDU/'
zip_paths = ['Dataset.zip', 'Dataset-2.zip', 'Hyps.zip']
destination_path = '/content/data/'

# Helper function to extract zip files
def extract_zip(zip_file, destination):
    try:
        with zipfile.ZipFile(zip_file, 'r') as zip_ref:
            zip_ref.extractall(destination)
        print(f"Extraction successful: {zip_file}")
    except Exception as e:
        print(f"Error extracting {zip_file}: {e}")

# Loop through and extract each zip file
for zip_file in zip_paths:
    full_zip_path = os.path.join(drive_path, zip_file)
    extract_zip(full_zip_path, destination_path)
```

```python
import cv2
import os
import matplotlib.pyplot as plt

def parseLabelFileContents(fileContents):
    lines = fileContents.split("\n")
    splitLines = [line.split() for line in lines if line != ""]
    return splitLines

def getRectangleBoxCoordinates(labels, imgWidth):
    return [(int(float(label[1]) * imgWidth),
             int(float(label[2]) * imgWidth),
             int(float(label[3]) * imgWidth),
             int(float(label[4]) * imgWidth))
            for label in labels]

def getRectangleData(coords):
    if coords is None:
        return None
    x, y, width, height = coords
    half_width = width // 2
    half_height = height // 2
    return (x - half_width, y - half_height), (x + half_width, y + half_height)

def drawRectangleToImage(coordinates, image):
    coordinatesTuple = getRectangleData(coordinates)
    image = cv2.rectangle(image, coordinatesTuple[0], coordinatesTuple[1], (0, 255, 0), 2)

nrows, ncols = 3, 3
fig = plt.gcf()
fig.set_size_inches(ncols * 4, nrows * 4)

filenamesOfLabelsToShow = NamesOfLabelsToShow
filenamesOfImagesToShow = NamesOfImagesToShow

for i, labelFilename in enumerate(filenamesOfLabelsToShow):
    sp = plt.subplot(nrows, ncols, i + 1)
    sp.axis("Off")


    img = cv2.imread(os.path.join("data/dataset/dataset/images/train", filenamesOfImagesToShow[i]))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    with open(os.path.join("data/dataset/dataset/labels/train", labelFilename)) as f:
        labelFileContent = f.read()
    listOfCoordinates = getRectangleBoxCoordinates(parseLabelFileContents(labelFileContent), img.shape[1])

    if listOfCoordinates is not None:
        for coordinates in listOfCoordinates:
            drawRectangleToImage(coordinates, img)
    plt.imshow(img)

plt.savefig("DataDisplay.png", bbox_inches="tight")
plt.show()
```
